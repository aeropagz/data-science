```python
data = pd.read_csv("linreg_data.csv")[["X", "Y"]].values
x = data[:, 0]
X = np.vstack([np.ones_like(x), x]).T
y = data[:, 1]

Bs = np.linalg.inv(X.T@X) @ (X.T @ y)
lin = np.linspace(-3, 10, 100) 
preds = np.vstack([np.ones_like(lin), lin]).T @ Bs

plt.scatter(x, y)
plt.plot(lin, preds, color="red")
plt.title("linear")
Bs
```

```python
from scipy.stats import linregress

slope, intercept, r_value, p_value, std_err = linregress(x, y)

plt.scatter(x, y)
plt.plot(x, intercept + x * slope, color="red")
```
