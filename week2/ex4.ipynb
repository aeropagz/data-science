```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

X = np.array([[1, 2.5], [1, 2.0], [1,2.3], [1, 2.4]])
yT = np.array([6.0, 5.1, 5.7, 5.6])

x1 = np.linalg.inv(X.T @ X)
y1 = (X.T @ yT.T)


x1, y1, x1 @ y1
```

```python
data = pd.read_csv("linreg_data.csv")[["X", "Y"]].values
x = data[:, 0]
X = np.vstack([np.ones_like(x), x]).T
y = data[:, 1]

Bs = np.linalg.inv(X.T@X) @ (X.T @ y)
lin = np.linspace(-3, 10, 100) 
preds = np.vstack([np.ones_like(lin), lin]).T @ Bs


plt.scatter(x, y)
plt.title('Original Data')
plt.show()

plt.scatter(x, y)
plt.plot(lin, preds, color="red")
plt.title("linear")
```
